### 2. **Handle Missing Values:** Identify and address missing values to ensure data completeness.

```python
df.info()
```

```python
df['Age'].fillna(df['Age'].median(), inplace = True) #Fills all the missing values of age with the median of the Age coloumn
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True) #Fills all the missing values of Embarked with the value that appears most often in Embarked coloumn
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.info()
```

### 3. **Feature Scaling:** Normalize or standardize numerical features for consistent scaling.

```python
df[['Age', 'Fare']]
```

#### Standardization

```python
scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df[['Age', 'Fare']]
```

#### Normalization

```python
min_max_scaler = MinMaxScaler()
df[['Age', 'Fare']] = min_max_scaler.fit_transform(df[['Age', 'Fare']])
df[['Age', 'Fare']]
```
